```python
from txtai.embeddings import Embeddings
import json
```

```python
embeddings = Embeddings({
    
    "path": "sentence-transformers/all-MiniLM-L6-v2"
})
```

```python
# Open the JSON file and load its content into a list (Change the path to the JSON file accordingly)
with open("C:\\Studies\\SoftEng2\\AISearchEngine\\Backend\\Dataset.json", "r") as f:
    data = json.load(f)

# Now 'data' contains a list of dictionaries, each representing a record

# To get the number of records in the JSON file, you can simply use len(data)
number_of_records = len(data)
print(f'Number of records: {number_of_records}')
```

```python
data[1]
```

```python
# Create an empty list to store the Abstract values
abstract_values = []

# Iterate through a dataset of text records
for record in data:
    # Extract the "Abstract" field from each record and append it to the abstract_values list
    abstract_value = record.get("Abstract")
    abstract_values.append(abstract_value)

# Access the Abstract value at a specific index (e.g., index 4)
abstract_value_4 = abstract_values[4]

# Print the Abstract value at index 4
print(f"Abstract value for record at index 4: {abstract_value_4}")
```

```python
embeddings.index(abstract_values)
```

```python
res = embeddings.search("Study about Convolutional Neural Networks and it must be year 2015", 10)
```

```python
for item in res:
    print(item)
```

```python
res = embeddings.search("Computer Science especially JAVA", 5)

for r in res:
    record = data[r[0]]
    
    # Print each field on a new line
    text = "\n".join(f"{key}: {value}" for key, value in record.items())
    print(f"Text:\n{text}")
    print(f"Similarity: {r[1]}")
    print()
```
